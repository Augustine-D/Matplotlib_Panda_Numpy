```python
#legend outside graph
plt.figure(figsize=(8,5))
plt.title('Gas prices over Time (in USD) \n')

plt.plot(gas.Year, gas.USA, label='USA' , marker='.', color='blue')
plt.plot(gas.Year, gas.Canada, label='Cananda', marker='.', color='green')
plt.plot(gas['Year'], gas['South Korea'], label='South Korea',marker='.', color='red')
plt.plot(gas['Year'], gas.Australia, label='South Korea',marker='.', color='yellow')

plt.xticks(gas.Year[::3])
plt.xlabel('\n\n Years')
plt.ylabel('\n US Dollars')
plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas prices over Time (in USD) \n')

#plt.plot(gas.Year, gas.USA, label='USA' , marker='.', color='blue')
#plt.plot(gas.Year, gas.Canada, label='Cananda', marker='.', color='green')
#plt.plot(gas['Year'], gas['South Korea'], label='South Korea',marker='.', color='red')
#plt.plot(gas['Year'], gas.Australia, label='South Korea',marker='.', color='yellow')
for country in gas:
    if country !='Year':
        plt.plot(gas.Year, gas[country], marker='.')
plt.xticks(gas.Year[::3])
plt.xlabel('\n\n Years')
plt.ylabel('\n US Dollars')
plt.legend()
plt.show()
```

```python
#Get specific countries
plt.figure(figsize=(8,5))
plt.title('Gas prices over Time (in USD) \n')

#plt.plot(gas.Year, gas.USA, label='USA' , marker='.', color='blue')
#plt.plot(gas.Year, gas.Canada, label='Cananda', marker='.', color='green')
#plt.plot(gas['Year'], gas['South Korea'], label='South Korea',marker='.', color='red')
#plt.plot(gas['Year'], gas.Australia, label='South Korea',marker='.', color='yellow')
countries_to_look_at = ['USA', 'Canada']
for country in gas:
    if country in countries_to_look_at:
        plt.plot(gas.Year, gas[country], marker='.')
plt.xticks(gas.Year[::3])
plt.xlabel('\n\n Years')
plt.ylabel('\n US Dollars')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.title('Gas prices over Time (in USD)\n', fontdict={'fontweight':'bold', 'fontsize':14} )

plt.plot(gas.Year, gas.USA, label='USA' , marker='.', color='blue')
plt.plot(gas.Year, gas.Canada, label='Cananda', marker='.', color='green')
plt.plot(gas['Year'], gas['South Korea'], label='South Korea',marker='.', color='red')
plt.plot(gas['Year'], gas.Australia, label='South Korea',marker='.', color='yellow')

plt.xticks(gas.Year[::3])
plt.xlabel('\n\n Years')
plt.ylabel('\n US Dollars')
plt.legend()
plt.show()
```

```python
#add another year
#Get specific countries
plt.figure(figsize=(8,5))
plt.title('Gas prices over Time (in USD)\n', fontdict={'fontweight':'bold', 'fontsize':14} )

plt.plot(gas.Year, gas.USA, label='USA' , marker='.', color='blue')
plt.plot(gas.Year, gas.Canada, label='Cananda', marker='.', color='green')
plt.plot(gas['Year'], gas['South Korea'], label='South Korea',marker='.', color='red')
plt.plot(gas['Year'], gas.Australia, label='South Korea',marker='.', color='yellow')

plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel('\n\n Years')
plt.ylabel('\n US Dollars')
plt.legend()
plt.show()
```

```python
#saving graph
plt.figure(figsize=(8,5))
plt.title('Gas prices over Time (in USD)\n', fontdict={'fontweight':'bold', 'fontsize':14} )

plt.plot(gas.Year, gas.USA, label='USA' , marker='.', color='blue')
plt.plot(gas.Year, gas.Canada, label='Cananda', marker='.', color='green')
plt.plot(gas['Year'], gas['South Korea'], label='South Korea',marker='.', color='red')
plt.plot(gas['Year'], gas.Australia, label='South Korea',marker='.', color='yellow')

plt.xticks(gas.Year[::3].tolist()+[2011])
plt.xlabel('\n\n Years')
plt.ylabel('\n US Dollars')
plt.legend()
#saving graph
plt.savefig('gas_price_figure.png', dpi=300)
plt.show()
```
